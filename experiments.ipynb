```python
import numpy as np
import matplotlib.pyplot as plt
import time

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

sample_dataset = [20,100,1000,10000]

for n in sample_dataset:
    noise = np.random.normal(0,2,n)
    print(noise)
    #dataset
    X = np.linspace(0, 10, n)
    y = 2*X + 1 + noise

    X = X.reshape(-1,1)

    model=LinearRegression()

    start = time.time()

    model.fit(X,y)

    end = time.time()

    predictions = model.predict(X)

    mse = mean_squared_error(y, predictions)
    r2 = r2_score(y, predictions)

    print(f"{n:10}{end-start:<15.6f}{mse:<15.8f}{r2:.6f}")
    

    plt.figure(figsize=(8,6))
    plt.scatter(X,y , label="datapoints", color='red')
    plt.plot(X, predictions , label="regressionline")
    plt.title(f"Dataset size = {n}")
    plt.ylabel("Y")
    plt.xlabel("X")
    plt.legend()
    plt.grid(True)
    plt.show()

```

```python

import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    r2_score
)

noise_levels = [5, 20, 50, 100]

for noise_std in noise_levels:

    # Dataset
    X = np.linspace(0,10,100)

    noise = np.random.normal(0, noise_std, 100)

    y = 2*X + 1 + noise

    X = X.reshape(-1,1)

    # Train
    model = LinearRegression()

    model.fit(X,y)

    predictions = model.predict(X)

    # Metrics
    mse = mean_squared_error(y,predictions)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y,predictions)
    r2 = r2_score(y,predictions)

    print("\n"+"="*50)
    print(f"Noise = {noise_std}")
    print("="*50)

    print(f"MSE  : {mse:.3f}")
    print(f"RMSE : {rmse:.3f}")
    print(f"MAE  : {mae:.3f}")
    print(f"R²   : {r2:.3f}")

    # Plot
    plt.figure(figsize=(7,5))

    plt.scatter(X,y,color="red",label="Data")

    plt.plot(X,predictions,color="blue",label="Regression Line")

    plt.title(f"Noise = {noise_std}")

    plt.xlabel("X")

    plt.ylabel("Y")

    plt.legend()

    plt.grid(True)

    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

X = np.linspace(-5,5,100)

y = X**2

plt.figure(figsize=(7,5))
plt.scatter(X,y,color="red")
plt.title("Original Dataset")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
X_linear = X.reshape(-1,1)

model = LinearRegression()

model.fit(X_linear,y)

predictions = model.predict(X_linear)

plt.figure(figsize=(7,5))

plt.scatter(X,y,color="red",label="Actual")

plt.plot(X,predictions,color="blue",linewidth=3,label="Linear Regression")

plt.title("Linear Regression on Polynomial Data")

plt.legend()

plt.grid(True)

plt.show()

print("MSE :",mean_squared_error(y,predictions))
print("R² :",r2_score(y,predictions))
```

```python
poly = PolynomialFeatures(degree=2)

X_poly = poly.fit_transform(X_linear)

poly_model = LinearRegression()

poly_model.fit(X_poly,y)

poly_predictions = poly_model.predict(X_poly)

plt.figure(figsize=(7,5))

plt.scatter(X,y,color="red",label="Actual")

plt.plot(X,poly_predictions,color="green",linewidth=3,label="Polynomial Regression")

plt.title("Polynomial Regression")

plt.legend()

plt.grid(True)

plt.show()

print("MSE :",mean_squared_error(y,poly_predictions))
print("R² :",r2_score(y,poly_predictions))
```
